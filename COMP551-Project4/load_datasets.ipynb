```
# Helper functions for manipulating .tar.gz files 
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='UTF-8')
    return dict

def unzip(file):
  import tarfile
  tar = tarfile.open(file, "r:gz")
  tar.extractall()
  tar.close()
```

```
# Unzip CIFAR10 and CIFAR100 datasets
unzip(cifar10_path)
unzip(cifar100_path)
```
